```python
import pandas as pd

# Load the datasets
countries_df = pd.read_csv('data/gapminder/country_continent.csv')
gdp_df = pd.read_csv('data/gapminder/gdp_pcap.csv')
hdi_df = pd.read_csv('data/gapminder/hdi_human_development_index.csv')
life_expectancy_df = pd.read_csv('data/gapminder/lex.csv')
population_df = pd.read_csv('data/gapminder/pop.csv')

# Reshaping the GDP, HDI, Life Expectancy, and Population dataframes to have a 'year' column
gdp_long_df = gdp_df.melt(id_vars=['country'], var_name='year', value_name='gdp')
hdi_long_df = hdi_df.melt(id_vars=['country'], var_name='year', value_name='hdi')
life_expectancy_long_df = life_expectancy_df.melt(id_vars=['country'], var_name='year', value_name='life_expectancy')
population_long_df = population_df.melt(id_vars=['country'], var_name='year', value_name='population')

# Converting 'year' column to integer for proper merging
gdp_long_df['year'] = gdp_long_df['year'].astype(int)
hdi_long_df['year'] = hdi_long_df['year'].astype(int)
life_expectancy_long_df['year'] = life_expectancy_long_df['year'].astype(int)
population_long_df['year'] = population_long_df['year'].astype(int)

# Merging the datasets
merged_df = countries_df.merge(life_expectancy_long_df, on='country', how='left')
merged_df = merged_df.merge(population_long_df, on=['country', 'year'], how='left')
merged_df = merged_df.merge(gdp_long_df, on=['country', 'year'], how='left')
merged_df = merged_df.merge(hdi_long_df, on=['country', 'year'], how='left')

# Reordering columns to match the required format
merged_df = merged_df[['country', 'continent', 'year', 'life_expectancy', 'population', 'gdp', 'hdi']]
merged_df = merged_df[merged_df['year'] <= 2023]

# Save the final dataset to a CSV file
merged_df.to_csv('data/gapminder/merged_dataset.csv', index=False)

merged_df.head()

merged_df = pd.read_csv('data/gapminder/merged_dataset.csv')

def convert_value(value):
    """
    Converts a population value from string format with suffixes (k, M, B) to an integer.
    For example: '30k' -> 30000, '2.01M' -> 2010000, '3.4B' -> 3400000000
    """
    if pd.isnull(value):
        return None
    if value[-1] == 'k':
        return int(float(value[:-1]) * 1e3)
    elif value[-1] == 'M':
        return int(float(value[:-1]) * 1e6)
    elif value[-1] == 'B':
        return int(float(value[:-1]) * 1e9)
    else:
        return int(value)

# Apply the conversion to the 'population' column
merged_df['population'] = merged_df['population'].apply(convert_value)
merged_df['gdp'] = merged_df['gdp'].apply(convert_value)

# Save the updated dataset to a new CSV file
updated_csv_path = 'data/gapminder/merged_dataset.csv'
merged_df.to_csv(updated_csv_path, index=False)

merged_df.head()
```
